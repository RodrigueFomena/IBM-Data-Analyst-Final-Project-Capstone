# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
QUERY = """
SELECT * FROM master
"""

df = pd.read_sql_query(QUERY,conn)
print("The number of columns is:", df.shape[1], "\nAnd the names of columns are:\n", df.columns)
print("Number of rows:", df.shape[0])
```

```python
#Plot a histogram of ConvertedComp

import matplotlib.pyplot as plt
import seaborn as sns

plt.hist(df.ConvertedComp)
plt.xlabel("ConvertedComp")
plt.ylabel("Renspondent Count")
plt.title("Distribution of ConvertedComp");
```

### Box Plots


Plot a box plot of `Age.`


```python
#Plot a box plot of Age

print("Maximum Age:", df.Age.max())
sns.boxplot(df.Age)
plt.ylabel("Age")
plt.title("Age Distribution & Outliers");
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
#Create a scatter plot of Age and WorkWeekHrs
sns.scatterplot(x=df.Age, y=df.WorkWeekHrs, data=df)
plt.xlabel("Age")
plt.ylabel("WorkWeekHrs")
plt.title("correlation between Age and WorkWeekHrs");

```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
#Create a bubble plot of WorkWeekHrs and CodeRevHrs, use Age column as bubble size

sns.scatterplot(x=df.WorkWeekHrs, y=df.CodeRevHrs, data=df, size=df.Age, hue=df.Age, alpha=0.8, sizes=(0,400))
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.title("correlation between WorkWeekHrs and CodeRevHrs by Age");
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.

```python
#list of all tables in sqlite_master

QUERRY1 = """SELECT name as Table_Name FROM 
sqlite_master WHERE
type = 'table'"""

pd.read_sql_query(QUERRY1, conn)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
#Create a DataFrame for the stacked chart

df_stacked = df[["Age", "WorkWeekHrs", "CodeRevHrs"]]
df_stacked = df_stacked[(df_stacked["Age"]>=30) & (df_stacked["Age"]<=35)]
df_stacked = df_stacked.groupby("Age").median()
df_stacked

```

```python
#Create the stacked chart

df_stacked.plot(kind='bar', stacked=True)
plt.xlabel("Age")
plt.ylabel("Median Hours")
plt.title("Median WorkWeekHrs and CodeRevHrs for Age Group 30-35")
plt.legend( bbox_to_anchor=(1, 1));
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
#Create a DataFrame for the line chart

df_line = df[["Age", "ConvertedComp"]]
df_line = df_line[(df_line["Age"]>=45) & (df_line["Age"]<=60)]
df_line = df_line.groupby("Age").median()
df_line
```

```python
#Create the line chart
df_line.plot(kind="line", figsize=(10,5))
plt.ylabel("ConvertedComp")
plt.title("Median Salary by Age");
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
#Create a DataFrame for the bar chart

df_bar = df.MainBranch.value_counts()
df_bar
```

```python
#Create a horizontal bar chart using column MainBranch

df_bar.plot(kind="barh")
plt.title('MainBranch Distribution')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.tick_params(axis='y')
plt.legend()
```

```python
# Another code to create a bar chart and to display the values on the y-axis in blocs

# import the textwrap library in Python
import textwrap

# Create the horizontal bar chart using column MainBranch
fig, ax = plt.subplots()
ax.barh(df_bar.index, df_bar.values)
ax.set_title('MainBranch Distribution')
ax.set_xlabel('Count')
ax.set_ylabel('MainBranch')
ax.tick_params(axis='y', labelsize=8)

# Displaying values in blocks on the y-axis
labels = [textwrap.fill(text, 10) for text in df_bar.index] # Split long texts into blocks
ax.set_yticklabels(labels, multialignment='left');

```
